```python
import pandas as pd
import re
import numpy as np
import matplotlib.pyplot as plt
from textblob import TextBlob
import yfinance as yf
import plotly.express as px
# silence warnings
import warnings
warnings.filterwarnings('ignore')
# pio.renderers.default = 'browser'
import os
```

```python
def fetch_stock_prices():
    # Define the directory containing the CSV files
    csv_directory = '../../data/yfinance_data/'

    # List of tickers (assuming the CSV files are named like 'AAPL.csv', 'MSFT.csv', etc.)
    tickers = ['AAPL', 'MSFT', 'GOOG']

    # Dictionary to hold DataFrames for each ticker
    stock_data = {}

    for ticker in tickers:
        file_path = os.path.join(csv_directory, f'{ticker}.csv')
        df = pd.read_csv(file_path)
        df['Date'] = pd.to_datetime(df['Date'])
        df.set_index('Date', inplace=True)
        stock_data[ticker] = df  # Store DataFrame in the dictionary

    # Example: Print the first few rows of the MSFT data
    print(stock_data['MSFT'].head())
    
    hist = stock_data['MSFT'].head()
    print(hist.head())
    print(hist.tail())
    fig = px.line(hist, x = hist.index, y = [hist.Close,hist.Open,hist.High,hist.Low], title='Microsoft Stock Analysis')
    fig.show()

    stock_data['MSFT'].plot(y='Close', title='Microsoft Stock Analysis', figsize=(10, 6))
    plt.xlabel('Date')
    plt.ylabel('Stock Price')
    plt.show()
```

```python
def eda():
    fetch_stock_prices()
```
